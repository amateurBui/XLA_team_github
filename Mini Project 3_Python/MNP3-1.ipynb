```python
import cv2  # OpenCV để xử lý ảnh
from PIL import Image  # Thư viện PIL để xử lý ảnh dạng RGB
import numpy as np  # Thư viện toán học cho xử lý ma trận

# Khai báo đường dẫn ảnh (sử dụng r"..." để tránh lỗi escape sequence)
filehinh = r"D:\Chuyen nganh\Computer Vision\XLA_team_github\XLA_team_github\bird_small.jpg"

# Đọc ảnh màu dùng OpenCV
img = cv2.imread(filehinh, cv2.IMREAD_COLOR)
if img is None:
    print("Lỗi: Không thể đọc ảnh từ đường dẫn!")
    exit()

# Đọc ảnh bằng thư viện PIL
imgPIL = Image.open(filehinh)

# Tạo ảnh kết quả có cùng kích thước
binary = Image.new(imgPIL.mode, imgPIL.size)

# Lấy kích thước ảnh
width, height = imgPIL.size

# Thiết lập ngưỡng
Nguong = 120

# Duyệt qua từng pixel để chuyển đổi ảnh sang ảnh nhị phân
for x in range(width):
    for y in range(height):
        # Lấy giá trị R, G, B
        R, G, B = imgPIL.getpixel((x, y))

        # Tính mức xám bằng công thức Luminance
        gray = int(0.2126 * R + 0.7152 * G + 0.0722 * B)  # Dùng int thay vì np.uint8

        # Xác định giá trị nhị phân
        if gray < Nguong:
            binary.putpixel((x, y), (0, 0, 0))  # Pixel đen
        else:
            binary.putpixel((x, y), (255, 255, 255))  # Pixel trắng

# Chuyển ảnh từ PIL sang OpenCV để hiển thị
nhiphan = np.array(binary)

# Hiển thị ảnh
cv2.imshow("Original Image", img)
cv2.imshow("Binary Image", nhiphan)

# Bấm phím bất kỳ để đóng cửa sổ hiển thị
cv2.waitKey(0)
cv2.destroyAllWindows()
```
